# Precificação Inteligente no ramo Imobiliário (Preparação dos Dados)



```python
import pandas as pd
```

```python
dados = pd.read_json('Dados/dados_hospedagem.json') # Apresenta os dados de forma aninhada
  
dados.head()
```

```python
# Realizar a normalização dos dados
dados = pd.json_normalize(dados['info_moveis'])
dados.head()
```

- Muitos dados estão agrupados em listas, e para que seja possível trabalhar com os mesmos é importante fazer as separações
-  Utilização do Método Explode ( separa os elementos para transformar os mesmos em linhas)

```python
# Obter o nome das colunas do dataframe
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(column=colunas[3:])
dados
```

```python
dados.reset_index(inplace=True, drop=True) # Comando drop permite criar uma nova coluna com os indices antigos
dados.head()
```

```python
dados.info()
```

# Convertendo Dados Numéricos
- Método astype: converte colunas para um tipo especifico de dados

```python
import numpy as np
```

```python
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']

dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
dados.info()
```

## Tratando dados númericos: convertendo para tipo floata

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
#dados['taxa_deposito'] = dados['taxa_deposito'].str.replace('$', '').str.replace(',', '').astype(np.float64)
```

- Método apply lambda: Aplica uma função específica a cada elemento

```python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
# Converter múltiplas colunas diferentes (utilza o applymap)
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```

## Tratando Dados Textuais
- Criação de tokenização: dividir o texto em unidades menores
  - Ex: ['Internet', 'Wireless']
  - Outra etapa seria remover as setpwords

```python
# Transformar o texto para minusculo

dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
dados.head()
```

## Removendo caracteres com regex (expressão regular)

```python
dados['descricao_local'][3169]
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']',' ',regex=True) # Sinal ^ nega quais caracteres não serão considerados
```

```python
# Removendo os hifen fora das palavras
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)',' ',regex=True) # Verifica se não possui caractere antes do hifen e depois também
```

# Tokenização dos textos

```python
dados['descricao_local'] = dados['descricao_local'].str.split(' ')
```

```python
dados.head()
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"','', regex=True)
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```

```python

dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True).replace('(?<!\w)-(?!\w)',' ',regex=True).str.lower()
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split(' ')

dados.head()
```
