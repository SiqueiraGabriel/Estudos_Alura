# Análise exploratória de dados

## Qual o valor médio de aluguel por tipo de imóvel?

```python
dados.head()
```

```python
# Obter o valor médio da coluna
dados['Valor'].mean()
```

```python
# Realizar o agrupamento dos dados (usado junto com outro método de agregação)
dados.groupby(by='Tipo').mean(numeric_only=True) # Calcular apenas as colunas numericas (obrigatório desta forma)
```

```python
# Obter a média a partir de agrupamento
dados.groupby(by='Tipo')['Valor'].mean()
```

```python
# Fazer o agrupamento e a ordenção dados dados
dados.groupby(by='Tipo')[['Valor']].mean().sort_values(by='Valor')
```

```python
# Plotar um gráfico
df_preco_tipo = dados.groupby(by='Tipo')[['Valor']].mean().sort_values(by='Valor')

df_preco_tipo.plot(kind='barh', figsize=(14,10), color='blue')
```

## Removendo os imóveis comerciais

```python
# Mostrar dados de tipo único
dados['Tipo'].unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala', 
                      'Prédio Inteiro', 'Loja/Salão', 
                      'Galpão/Depósito/Armazém', 
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
                                            
```

```python
# Metodo que permite selecionar linhas do datframe com base em colunas especificas por meio de expressão
# QUando na expressão deseja passar uma variável, deve-se colocar um @ antes
dados.query('@imoveis_comerciais in Tipo')
```

```python
dados.query('@imoveis_comerciais not in Tipo')
```

```python
df = dados.query('@imoveis_comerciais not in Tipo')
df
```

```python
df.Tipo.unique()
```

```python
df_preco_tipo = df.groupby(by='Tipo')[['Valor']].mean().sort_values(by='Valor')

df_preco_tipo.plot(kind='barh', figsize=(14,10), color='blue')
```

## Qual o percentual de cada tipo de imóvel na nossa base de dados?

```python
df.Tipo.unique()
```

```python
# Contar a quantidade de vezes que cada elemento aparece
df.Tipo.value_counts()
```

```python
# Contar a quantidade de vezes que cada elemento aparece no formato porcentual
df.Tipo.value_counts(normalize=True)
```

```python
# Transformar o resultado para uma tabela
df.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
df_percentual_tipo = df.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
df_percentual_tipo.plot(kind='bar', figsize=(14,10), color='blue', xlabel='Tipos', ylabel='Percentual')
```

## Desafios

1) Calcular a média de quartos dos imóveis;

2) Conferir quantos bairros únicos existem na nossa base de dados;

3) Analisar quais bairros possuem a média de valor de aluguel mais elevadas;

4) Criar um gráfico de barras horizontais que apresente os 5 bairros com as médias de valores de aluguel mais elevadas.

```python
# 1) Calcular a média de quartos dos imóveis;
df.Quartos.mean()
```

```python
# 2) Conferir quantos bairros únicos existem na nossa base de dados;
bairos_unicos = df.Bairro.unique()

print(bairos_unicos)

print(f"Quantidade de bairos unicos {len(bairos_unicos)}")
```

```python
# 2.1 Metodo alternativo para obter a quantidade de elementos únicos
df.Bairro.nunique()
```

```python
# 3) Analisar quais bairros possuem a média de valor de aluguel mais elevadas;
df.groupby(by='Bairro').Valor.mean().sort_values(ascending=False).to_frame().head(n=10)
```

```python
#4) Criar um gráfico de barras horizontais que apresente os 5 bairros com as médias de valores de aluguel mais elevadas.

df_top_5Bairros_media_preco = df.groupby(by='Bairro').Valor.mean().sort_values(ascending=False).to_frame().head(n=5)

df_top_5Bairros_media_preco.plot(kind='barh', figsize=(10, 14), color='blue')
```
