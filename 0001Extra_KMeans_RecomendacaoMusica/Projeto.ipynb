## PCA e StandardScaller 

### Clusterização Por Genero

```python
dados_generos
```

```python
dados_generos1 = dados_generos.drop('genres', axis=1)
dados_generos1
```

```python
# Bibliotecas do sklearn para Pipeline
from sklearn.pipeline import Pipeline    #Realiza a criação de pipeline de forma automática, que corresponde a forma como vamos usar o Machine Learning
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
SEED = 1224             # Definir que os dados sejam aleatórios mas que seja possível replicar
np.random.seed(1224)

# Realizar a padronização dos dados
pca_pipeline = Pipeline([('scaler', StandardScaler()), ('PCA', PCA(n_components=2, random_state=SEED))]) # Lista dos modulos a ser utilizado para a criação da Pipeline
```

```python
genre_embedding_pca = pca_pipeline.fit_transform(dados_generos1)    # Treinamento com os dados de generos
projection = pd.DataFrame(columns=['x', 'y'], data=genre_embedding_pca) # 
```

```python
projection
```

### K-Means

```python
from sklearn.cluster import KMeans

kmeans_pca = KMeans(n_clusters=5, verbose=True, random_state=SEED) # Verbose: ver o que esta acontecendo
kmeans_pca.fit(projection)

dados_generos['cluster_pca'] = kmeans_pca.predict(projection) # Salvar o genero que cada cluster possui
projection['cluster_pca'] = kmeans_pca.predict(projection) # Salvar o genero que cada cluster possui
```

```python
projection
```

```python
# Dentro dos projections s
projection['generos'] = dados_generos['genres']
projection
```

### Plotando o Cluster

```python
fig = px.scatter(
    projection, x='x', y='y', color='cluster_pca', hover_data=['x', 'y', 'generos']
)
fig.show()
```

```python
# Avaliar a qualidade do cluster
pca_pipeline[1].explained_variance_ratio_.sum() # Taxa de quase 50% dos dados explicados
```
