```python
import pandas as pd
```

```python
sheet_id = '1ewVJUlqJcj8qw3exgtju8ZPG8zAT_O-PsonUJw0hpDs'
```

```python
# Para conseguir acessar a planilha, precisa-se adicionar o parâmetro '/gviz/tq' (API do Google) -
#  '?tqx=out:csv' especifica que os dados serão retornados em csv
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet'   
```

```python
dados_co2_sheets = pd.read_csv(url)
```
